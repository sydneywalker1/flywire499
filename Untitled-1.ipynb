```python
results = connections.groupby(['pre_root_id', 'post_root_id','neuropil'])['syn_count'].sum().reset_index()
```

```python
def third_order(neuron_data, connection_dataset,taste_neurons):
    for entry in neuron_data['second_order_cell_ID']:
        if entry in sugar_neurons['root_id']:
            pass
        else:
            merge = pd.merge(neuron_data[['second_order_cell_ID','label']], connections, left_on='second_order_cell_ID', right_on='pre_root_id', how = 'inner').query("syn_count >=5")
        merge = merge.drop(columns = 'pre_root_id')
        merge = merge.rename(columns={'post_root_id':'third_order_cell_ID','label':'second_order_cell_label'})
        def projection(neuropil):
            if neuropil == 'GNG':
                return 'local'
            else: 
                return 'outside SEZ'
        merge['location_of_connection'] = merge['neuropil'].apply(projection)
        return merge
```

```python
def function(result, taste_neurons, neuron_downstream, name):
    merge = pd.merge(result, taste_neurons['root_id'], left_on = 'pre_root_id', right_on = 'root_id', how = 'inner')
    merge = merge.drop(columns='root_id')
    merge2 = pd.merge(merge, neuron_downstream, left_on = 'post_root_id', right_on = 'root_id', how = 'inner')
    connections = merge2.groupby('post_root_id')['syn_count'].sum().reset_index()
    labels = pd.merge(connections, neuron_downstream,left_on = 'post_root_id', right_on = 'root_id', how = 'inner')
    labels = labels.drop(columns = 'root_id')
    labels.rename(columns={'input_synapses': 'total_#_input_synapses','syn_count':'#_input_' + name + '_synapses','post_root_id':'second_order_cell_ID'}, inplace=True)
    labels['% input from ' + name + ' GRNs'] = labels['#_input_' + name + '_synapses'] / labels['total_#_input_synapses'] * 100
    labels = labels.sort_values(by = '% input from ' + name + ' GRNs', ascending = False)
    return labels
```

```python
sugar_data = function(results, sugar_neurons, sugar_downstream, "sugar")
bitter_data = function(results, bitter_neurons, bitter_downstream, "bitter")
water_data = function(results, water_neurons, water_downstream, "water")
lowsalt_data = function(results, lowsalt_neurons, lowsalt_downstream, "lowsalt")
```

```python
def threshold(taste_third_order):
    groups = taste_third_order.groupby("second_order_cell_ID")
    taste_third_order["#_outside_connections"] = groups["location_of_connection"].transform(lambda x: (x == "outside SEZ").sum())
    taste_third_order["total_connections"] = groups["location_of_connection"].transform("count")
    taste_third_order["fraction_nonlocal"] = taste_third_order["#_outside_connections"] / taste_third_order["total_connections"]
    test = taste_third_order 
    test2 = test.drop_duplicates(subset = 'second_order_cell_ID').reset_index(drop=True)
    return test2
```

```python
sugar_threshold = threshold(sugar_third_order)
bitter_threshold = threshold(bitter_third_order)
water_threshold = threshold(water_third_order)
lowsalt_threshold = threshold(lowsalt_third_order)
```

```python
 def classify(taste_threshold):
    classification = pd.DataFrame(columns = ['second_order_cell_ID', 'projection','%_of_external_projections','nt_type'])
    unique_values = taste_threshold['second_order_cell_ID'].unique()
    for entry in unique_values:
        if (taste_threshold[taste_threshold['second_order_cell_ID'] == entry ]['fraction_nonlocal'] < 0.10 ).any():
            x = pd.DataFrame({'second_order_cell_ID': [entry], 'projection':'local','%_of_external_projections': np.mean(taste_threshold[taste_threshold['second_order_cell_ID'] == entry]['fraction_nonlocal']), 'nt_type': taste_threshold[taste_threshold['second_order_cell_ID'] == entry]['nt_type']})
            classification = pd.concat([classification,x])
        else:
            x = pd.DataFrame({'second_order_cell_ID': [entry], 'projection':'outside SEZ','%_of_external_projections':np.mean(taste_threshold[taste_threshold['second_order_cell_ID'] == entry]['fraction_nonlocal']),'nt_type': taste_threshold[taste_threshold['second_order_cell_ID'] == entry]['nt_type']})
            classification = pd.concat([classification,x])
    return classification
```

```python
sugar_second_order_projection = classify(sugar_threshold)
bitter_second_order_projection = classify(bitter_threshold)
water_second_order_projection = classify(water_threshold)
lowsalt_second_order_projection = classify(lowsalt_threshold)
```
